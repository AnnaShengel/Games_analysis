### 1. Работа с данными

Для начала импортируем все нужные для проекта библиотеки.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn
import pylab as pl
from scipy import stats as st
```

#### 1.1. Знакомтство с данными

Откроем нужный csv-файл  и сохраним в переменную data. Выведем первые строки датафрейма и общую информацию о файле.

```python
data = pd.read_csv('games.csv')
data.head()
```

```python
#выведем оющую информацию о датасете
data.info()
```

Чтобы в дальнейшем было удобнее работать, приведем названия столбцов в нижний регистр, избавимся от пропусков, где это возможно и поменяем тип данных, где нужно.

**1.2 Предобработка данных**

**Названия столбцов**

Приведем названия столбцов к нижнему регистру.

```python
data.columns = data.columns.str.lower()

#Посмотрим, что у нас получилось
data.head(3)
```

**Работа с пропусками**

Выведем количество пропусков для каждого столбца, а также долю пропусков.

```python
print(data.isna().sum())
data.isna().mean()*100
```

Есть незначительно количество пропусков в столбцах с названием игры, годом выпуска и жанром игры. Пропуски в этих столбцах мскорее всего случайные, так как эти данные должны присутствовать во всех играх. Пропуски в этих столбцах мы не можем ничем заменить и доля этих пропусков не больше 2%, поэтому удалим строки с пропусками в этих столбцах.

```python
data = data.dropna(subset=['name', 'year_of_release', 'genre'])
```

Как правило, критики оценивают только самые популярные игры, поэтому в этом столбце `critic_score` мы имеем больше всего пропусков. Пропуски в этом столбце оставим, не будем ни чем заменять.

В столбце `user_score`  пропуски также неслучайные. Многие игры из-за непопулярности могут и не иметь оценки пользователей либо количество отзывов слишком маленькое. Пропуски также оставим как есть, так как заменить их корректно мы ничем не можем.

В столбце `rating` представлен возрастной рейтинг, определенных ESRB. В датасете много пропусков, так как рейтинг ESRB предназначен для маркировки игр для США и Канада, а в файле присутствуют данные по играм по всему миру. Также ESRB был основан 1994 году, следовательно, игры выпущенные до 1994 не имеют такого рейтинга.

Для дальнейшего анализа нам важно будет учитывать и игры без рейтинга, поэтому заменим пропуски на заглушку с значением 


```python
pd.options.mode.chained_assignment = None
data['rating'] = data['rating'].fillna('Unknown')

#Проверим еще раз пропуски
print(data.isna().sum())
```

В столбце `user_score` сохранены оценки пользователей, но тип данных этого столбца - текстовой. Выведем уникальные значения данного столбца, чтобы понять какие данные хранятся в этом столбце.

```python
data['user_score'].unique()
```

Видим, что кроме оценок и пропусков есть значения `tbd`. Посмотрим количество строк с такими значениями.

```python
data.query('user_score == "tbd"')['user_score'].count()
```

Так как строк с значением `tbd` в столбце `user_score` достаточное количество и нет разъяснения данного вида оценки, заменим такие значения на пропуски.

```python
data['user_score'] = data['user_score'].replace('tbd', np.NaN)

#Проверим, увеличилось ли кол-во пропусков в столбце user_score
print(data.isna().sum())
```

Избавились от пропусков, где это возможно.

**Работа с типами данных**

В столбце `user_score` хранятся оценки пользователей, поэтому поменяем тип данных данного столбца на вещественные числа. Используем тип float32.

```python
data['user_score'] = data['user_score'].astype('float32')
```

В столбце `year_of_release` указан год выпуска, тип столбца вещественное число. Так как в данных указан только год выпуска, а не полная дата, преобразуем тип на целочисленные числа.

```python
data['year_of_release'] = data['year_of_release'].astype('int32')

#Выведем еще раз общую информацию о таблице
data.info()
```

**Работа с дубликатами**

Проверим датасет на наличие явных дубликатов. Перед этим приведем названия игр в нижний регистр, чтобы исключить наличие неявных дубликатов.

```python
data['name'] = data['name'].str.lower()
data.duplicated().sum()
```

Данные о продажах игр не имеют дубликатов.

#### 1.3 Суммарные продажи

Подсчитаем суммарные продажи во всех регионах и запишем их в отдельный столбец `total_sales`.

```python
data['total_sales'] = data['na_sales'] + data['eu_sales'] + data['jp_sales'] + data['other_sales']

#Выведем первые строки таблицы, чтобы проверить на корректность.
data.head()
```

**Выводы:**

На данном этапе мы проделали следующие шаги:
- познакомились с данными;
- для удобства привели названия столбцов в нижний регистр;
- провели работу с пропусками;
- провели работу с типами данных;
- проверили наличие дубликатов.


### 2. Исследовательский анализ данных

#### 2.1 Сколько игр выпускалось в разные годы

Посмотрим, сколько игр выпускалось в разные годы. Для этого сгруппируем данные по столбцу `year_of_release` и построим столбчатую диаграмму.

```python
data.groupby('year_of_release')['name'].count().plot(kind='bar')
pl.title('Количество выпущенных игр по годам')
pl.xlabel('Год выпуска')
pl.ylabel('Количество игр')
```

Начиная с 1992 года количество выпускаемых игр стремительно возрастало, это связно с тем, что в 1990-е годы компьютерные игры перешли от растровой графики к полностью полигональному 3D (ист. wikipedia.org).

Мировой экономический кризис в 2008 год не обошел мимо и игровую индустрию, по этой причине на графике можно заметить значительный спад, начиная с 2009 года. А вот с 2013 года игровая индустрия приходит в норму и начинает расти.

#### 2.2. Как менялись продажи по платформам

Изучим, как менялись продажи по платформам. Для этого создадим сводную таблицу  по общим продажам для каждой платорфмы и построим столбчатую диаграмму.

```python
(data
    .pivot_table(index='platform', values='total_sales', aggfunc='sum')
    .reset_index()
    .sort_values('total_sales', ascending=False)
    .plot(x='platform', y='total_sales', kind='bar', label='Общие продажи')
)
pl.title('Общие продажи по платформам за все годы')
pl.xlabel('Платформа')
pl.ylabel('Продажи, млн. копий')
```

Из диаграммы видно, что самыми популярными платформами за все время были PS2, X360, PS3, Wii, DS и PS.

Далее изучим, как менялись продажи для популярных платформ по годам.

```python
#Создадим лист с названиями популярных платформ
top_platform = ['PS', 'PS2', 'PS3', 'X360', 'Wii', 'DS']

#Для каждой платформы из списка построим графики
for platforms in top_platform:
    (data
        .query('platform in @platforms')
        .pivot_table(index=['year_of_release'], values='total_sales', aggfunc='sum')
        .reset_index()
        .plot(x='year_of_release', y='total_sales', title=platforms, kind='bar', xlim=(1980, 2016), label='Общие продажи')
    )
    pl.xlabel('Год выпуска')
    pl.ylabel('Продажи, млн. копий')
```

На графике платформы DS видим, что есть данные за 1985 год, хотя Nintendo DS была выпущена только в 2004 году. Выведем эту строку.

```python
data.query('platform == "DS" and year_of_release == 1985')
```

Вероятно это ошибка. Удалим эту строку, так как не удалось выявить корректную дату выпуска этой игры на платформе DS.    

```python
data = data.query('platform != "DS" and year_of_release != 1985')
```

Из графиков можно сделать следующие выводы:
- Цикл жизни популярных платформ в среднем составляет 10 лет.
- За 4-6 лет платформы достигают пика пользования.
- За 3-5 лет платформы теряют спрос и могут исчезнуть.

#### 2.3 Актуальный период
	
Исходя из вышеуказанных выводов, следует взять данные за последние 4 года, чтобы построить прогноз на 2017 года.

Создадим новую переменную `data_actual`, где передадим данные об играх, выпущенных позднее 2012 года .

```python
data_actual = data.query('year_of_release > 2012')
```

#### 2.4. Лидирующие платформы

Определим лидирующие платформы за актуальные период.  Для этого построим диаграмму продаж по платформам за актуальный период в общем и по годам.

```python
#Диаграмма продаж по платформам за актуальный период
(data_actual
    .pivot_table(index='platform', values='total_sales', aggfunc='sum')
    .reset_index()
    .sort_values('total_sales', ascending=False)
    .plot(x='platform', y='total_sales', kind='bar', title='Продажи по платформам за 2013-2016 гг.', label='Общие продажи')
)
pl.xlabel('Платформа')
pl.ylabel('Продажи, млн. копий')

#Диаграмма продаж по платформам в разрезе годов
(data_actual
    .pivot_table(index='platform', columns='year_of_release', values='total_sales', aggfunc='sum')
    .plot(kind='bar', figsize=(10, 5), grid=True, title='Продажи по платформам в разрезе годов')
)
pl.xlabel('Платформа')
pl.ylabel('Продажи, млн. копий')
```

Из графиков можно сделать следующие выводы:
- Лидируют по продажам за последние 4 года - PS4, PS3, XOne, 3DS и X360.
- Продажи на платформах PS3, 3DS и X360 хоть и лидируют, но наблюдается спад продаж за последние годы, т.е. через пару лет эти платформы скорее всего будут не актуальны.
- Потенциально прибыльными платформами являются PS4 и XOne. На этих платформах было больше всего продаж в 2016 году, также по графику видно, что наблюдается рост продаж за последние 3 года (не учитывая 2016 год).


#### 2.5 Диаграмма размаха

Построим диаграмму размаха - ящик с усами по глобальным продажам игр в разбивке по платформам.

```python
plt.figure(figsize=(11, 7))
seaborn.boxplot(data=data_actual, x='platform', y='total_sales')

pl.title('Диаграмма размаха продажи в разрезе платформ')
pl.xlabel('Платформа')
pl.ylabel('Продажи, млн. копий')
```

Самые популярные платформы имеют много выбросов. Чтобы легче читались графики, построим график без больших выбросов. 

```python
plt.figure(figsize=(11, 7))
seaborn.boxplot(data=data_actual, x='platform', y='total_sales').set(ylim=(0, 1.8))

pl.title('Диаграмма размаха продажи в разрезе платформ')
pl.xlabel('Платформа')
pl.ylabel('Продажи, млн. копий')
```

Если сравнивать третьи квартили, то лидирующими платформами являются X360, Wii, PS4, XOne и WiiU. Данные платформы также лидируют по медианным значениям продаж. Отстающими платформами как по третьим квартилям, так и по медианным значениям являются PSP и PSV.

#### 2.6 Влияние отзывов на продажи

Проверим, влияют ли отзывы пользователей и критиков на продажи. Возьмем для начала самую популярную платформу PS4, посчитаем коэффициент корреляции Пирсона и построим диаграмму рассеяния.

```python
#Выведем коэффициенты корреляции Пирсона между отзывами пользователей, оценками критиков и продажами.
print(data_actual.query('platform == "PS4"')[['user_score', 'critic_score', 'total_sales']].corr())

#Построим диаграммы рассеяния зависимости оценки пользователей и продаж
data_actual.query('platform == "PS4"').plot(
    x='user_score',
    y='total_sales',
    kind='scatter',
    title='Влияние отзывов пользователей на продажи для PS4'
)
pl.xlabel('Оценки пользователей')
pl.ylabel('Продажи, млн. копий')

#Построим диаграммы рассеяния зависимости оценки критиков и продаж
data_actual.query('platform == "PS4"').plot(
    x='critic_score',
    y='total_sales',
    kind='scatter',
    title='Влияние отзывов критиков на продажи для PS4'
)
pl.xlabel('Оценки критиков')
pl.ylabel('Продажи, млн. копий')
```

Коэффициент корреляции Пирсона между продажами и оценками пользователей получился отрицательным и близким к 0, т.е. это говорит о том, что оценки пользователей не влияют на продажи.

А вот коэффициент корреляции между продажами и оценками критиков составляет 0,56, т.е. есть некая связь между оценками критиков и продажами. По графику рассеяния видно, что чем выше оценка, тем больше продажи, либо же наоборот - чем популярнее игра, тем выше шанс, что оценка критиков будет выше.

Выше мы определили 5 популярных платформ за последние 4 года. Определим связь между продажами и оценками пользователей и критиков на остальных платформах.

```python
#Создадим список с топ платформами без PS4
top_platform_actual = ['PS3', 'XOne', '3DS', 'X360']

for platforms in top_platform_actual:
    print('Влияние отзывов на продажи на платформе', platforms)
    print(data_actual.query('platform in @platforms')[['user_score', 'critic_score', 'total_sales']].corr())
    
    data_actual.query('platform in @platforms').plot(x='user_score', y='total_sales', kind='scatter', title=platforms)
    pl.xlabel('Оценки пользоавтелей')
    pl.ylabel('Продажи, млн. копий')
    plt.show()
    
    data_actual.query('platform in @platforms').plot(x='critic_score', y='total_sales', kind='scatter', title=platforms)
    pl.xlabel('Оценки критиков')
    pl.ylabel('Продажи, млн. копий')
    plt.show()
```

На остальных популярных платформах коэффициент корреляции Пирсона между продажами и оценками пользователей так же почти равна 0, т.е. нет связи между этими данными.

Корреляция между продажами и оценками критиков на остальных платформах меньше чем значение на PS4. Если на PS4 есть некая связь между продажами и оценками критиков, то на других популярных платформах связь ниже. Но все же по графику видно, что чем выше оценка критиков, тем выше продажи.


#### 2.7 Распределение по жанрам

Изучим общее распределение игр по жанрам за актуальный период. Если мы возьмем сумму продаж по жанрам, из-за большого ассортиментов игр в определенных жанрах, результаты по прибыльности могут исказиться, поэтому сначала построим диаграмму размаха по продажам для каждого жанра. Так будут наглядно видны медианные продажи и выбросы.

```python
plt.figure(figsize=(10, 6))
seaborn.boxplot(data=data_actual, x='genre', y='total_sales')

pl.title('Диаграмма размаха продаж в разрезе жанров')
pl.xlabel('Жанр')
pl.ylabel('Продажи, млн. копий')
```

По диаграмме размаха видно, что в каждом жанре есть большие выбросы, поэтому для дальнейшего анализа лучше взять медианные значения продаж, так как средние значения очень чувствительны к выбросам. Выведем диаграмму размаху еще раз, но уже отрезав выбросы.

```python
plt.figure(figsize=(10, 6))
seaborn.boxplot(data=data_actual, x='genre', y='total_sales').set(ylim=(0, 3.5))

pl.title('Диаграмма размаха продаж в разрезе жанров')
pl.xlabel('Жанр')
pl.ylabel('Продажи, млн. копий')
```

Исходя из диаграммы можно сделать следующие выводы:
    
- Самыми популярными жанрами за 2013-2016 гг. являются Shooter, Sports и Platform.
    
- Самым непопулярным жанром является Adventure.
- Остальные жанры находятся примерно на одном уровне
    
Также построим диаграмму по медианным продажам для каждого жанра по годам.

#### 3.1 Популярные платформы по регионам

Для каждого региона определим топ-5 платформ по итогам продаж за последние 4 года и построим круговые диаграммы, чтобы увидеть доли продаж.

```python
#Зададим функцию, которая будет выводить продажи по платформам и круговую диаграмму на основе полученных данных
def top_5_platforms(title, column):
    print(title)
    data = (data_actual
        .pivot_table(index='platform', values=column, aggfunc='sum')
        .sort_values(by=column, ascending=False).head()
        .reset_index()   
        )
    #Посчитаем сумму продаж в остальынх платформах
    data_others = data_actual[column].sum() - data[column].sum()
    
    #Добавим в переменную data еще одну строку с суммой продаж в других платформах
    data = data.append({'platform': 'others', column: data_others}, ignore_index = True)
    print(data)
    
    #Построим кргувоую диаграмму получившейся таблицы
    data[column].plot.pie(
        subplots=True, labels=data['platform'],
        autopct='%.0f%%'
    )
    plt.show()

top_5_platforms(title='Топ-5 платформ в Северной Америке', column='na_sales')
top_5_platforms(title='Топ-5 платформ в Европе', column='eu_sales')
top_5_platforms(title='Топ-5 платформ в Японии', column='jp_sales')
```

В Европе каждый 3-ий игрок (36%) пользуются платформой PS4. Также в топ-5 попадают платформы PS3 (17%), XOne (13%), X360 (11%) и 3DS (8%), остальные 15% пользователей предпочитают другие платформы.

### 4. Проверка гипотезы

#### 4.1. Гипотеза о равенстве рейтингов платформ Xbox One и PC

Проверим гипотезу о том, что рейтинги платформ Xbox One и PC одинаковые.

Формулировка гипотезы:

**Нулевая гипотеза:** Средние пользовательские рейтинги платформ Xbox One и PC равны.

**Альтернативная гипотеза:**  Средние пользовательские рейтинги платформ Xbox One и PC не равны.

В данном случае будем проверять гипотезу о равенстве средних двух генеральных совокупностей. 

Пороговое значение сделаем 5%.


```python
#Создадим массивы XOne и PC для каждой платформы, передадим данные из столбца user_score,
#предварительно выбрав нужные платформы и убрав пропуски. 
XOne = data_actual.query('platform == "XOne"')['user_score'].dropna() 
PC = data_actual.query('platform == "PC"')['user_score'].dropna()

results = st.ttest_ind(XOne, PC, equal_var=False)

alpha = .05

print(results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу ')
```

Нулевую гипотезу не получилось отвергнуть, получается, что средние пользовательские рейтинги платформ Xbox One и PC действительно одинаковые.

#### 4.2. Гипотеза о равенстве рейтингв жанров Action и Sports

Проверим гипотезу о том, что средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») одинаковые.

**Нулевая гипотеза:** Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») равны.

**Альтернативная гипотеза:** Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») не равны.

Пороговое значение также сделаем 5%.

```python
#Создадим массивы action и sports для каждого жанра, передадим данные из столбца user_score,
#предварительно выбрав нужные жанры и убрав пропуски. 
action = data_actual.query('genre == "Action"')['user_score'].dropna()
sports = data_actual.query('genre == "Sports"')['user_score'].dropna()

results = st.ttest_ind(action, sports, equal_var=False)

alpha = .05

print(results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу ')
```

Нулевая гипотеза гласила о том, что средние пользовательские рейтинги жанров Action и Sports равны. По результатам проверки гипотезы, отвергли нулевую гипотезу, т. е. получается, что средние пользовательские рейтинги двух жанров разные.
